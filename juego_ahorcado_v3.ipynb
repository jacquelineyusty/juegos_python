```python
import random

# Saludamos al jugador y le damos la bienvenida.
print("*********************************************************************")
print("                   Â¡Bienvenid@ al juego del Ahorcado!")
print("               Un juego de: TRITURADORAS DE CÃ“DIGO, S.A.")
print("*********************************************************************\n")

# Interactuamos con el jugador solicitando respuesta activa para comenzar la partida
instrucciones_o_jugar = input("âŒ¨ï¸ Pulsa i para leer las instrucciones ğŸ“œo j para empezar a jugar â¯ï¸âŒ¨ï¸").lower() 

if instrucciones_o_jugar == 'i': # Si el jugador escoge 'i' le mostramos las instrucciones.
    print("Recordemos las instruccionesğŸ“œ:\n"

    "El objetivo del juego es adivinar una palabra en el menos nÃºmero de intentos posibles: \n"

      "- Primero se deben marcar tantas rayas como letras tenga la palabra a adivinar.\n"
      "- Se dibuja una horca y el jugador dirÃ¡ letras para adivinar la palabra.\n"
      "- Si la letra dicha por el jugador es acertada, se escribirÃ¡ en la raya de la posiciÃ³n correspondiente.\n"
      "- Si la letra dicha por el jugador no es acertada, se dibujarÃ¡ una parte del monigote.\n"
      "- Hay que conseguir adivinar la palabra antes de que se complete el cuerpo...\n\n"

      "                          Â¡o terminarÃ¡s â˜ ï¸ AHORCAD@!â˜ ï¸\n")

elif instrucciones_o_jugar == 'j': # Si el jugador escoge 'j' empezamos el juego.
    print("----------------------ğŸ–±ï¸Â¿PREPARAD@?ğŸ–±ï¸----------------------\n"
            "------------------------ğŸš€Â¿LIST@?ğŸš€----------------------\n"
            "-----------------ğŸ®Â¡COMIENZA EL JUEGO!ğŸ®-----------------\n")


vidas = 6 # Inicializamos contadores y variables del juego.
letras_adivinadas = set() # Creamos una variable de tipo set para asegurarnos de que no se repitan las letras dichas por el jugador.

# La maquina elige aleatoriamente la palabra secreta de entre una lista de palabras que hemos creado. 
lista_palabras = ["mar", "gato", "barco", "cielo", "pato", "erizo", "pez", "nube", "sol", "ola",
                  "pluma", "mundo", "libro", "plan", "sopa", "tecla", "feliz", "rosa", "punto", "nariz",
                  "campo", "lago", "luz", "duro", "menta", "pan", "reloj", "flor", "gris", "azul",
                  "submarino", "nadie", "cruel", "diez", "guisante", "copia", "hueco", "tapiz", "pies", "violeta",
                  "montar", "ley", "ave", "mano", "juego", "kilo", "vino", "salud", "fuego", "queso"]

palabra_secreta = random.choice(lista_palabras)
set_letras_palabra_secreta = set(palabra_secreta)

# Establecemos las variables que usaremos 
lista_palabra_usuario = list("_"*len(palabra_secreta))
palabra_oculta = " ".join(lista_palabra_usuario)
set_letras_acertadas = set()
set_letras_falladas = set()
horca = [  
    "  +-----+  ",
    "  |     |  ",
    "  |        ",
    "  |        ",
    "  |        ",
    "  |        ",
    "==========="
] #  Creamos una lista que contiene las lÃ­neas que representan la horca en su estado inicial, sin ningÃºn monigote dibujado

# Definimos una funciÃ³n para actualizar el dibujo de la horca
def actualizar_horca(vidas):
    if vidas <= 5:
        horca[2] = "  |     O  "
    if vidas <= 4:
        horca[3] = "  |     |  "
    if vidas <= 3:
        horca[3] = "  |    /|  "
    if vidas <= 2:
        horca[3] = "  |    /|\\"
    if vidas <= 1:
        horca[4] = "  |    /   "
    if vidas == 0:
        horca[4] = "  |    / \\"

# Mostramos los espacios de las letras que el usuario tiene que adivinar con guiones bajos en pantalla.
print(f"Acabo de pensar una palabra de {len(palabra_secreta)} letras.")    
print (f"Â¿SerÃ¡s capaz de acertarla, adivinando sus letras?ğŸ§\n")
print({palabra_oculta})


# Establecemos el primer bucle while para permitir que se puedan jugar mÃºltiples partidas
estado_programa = True
while True:
    # El programa se desarrolla en un bucle que se ejecuta mientras el usuario tenga vidas y no haya adivinado la palabra.
    start = vidas > 0 and set_letras_acertadas != set_letras_palabra_secreta 

    while start:

        letra = input("Introduce una letra").lower() # Pedimos al jugador que introduzca una letra

        if letra in palabra_secreta: 
            print("Correctoâœ…, Â¡tu letra estÃ¡ en la palabra!ğŸ‘")
            posicion_letra_acertada = palabra_secreta.index(letra) # Averiguamos la posiciÃ³n donde se encuentra la letra acertada.
            lista_palabra_usuario[posicion_letra_acertada]= letra  # Sustituimos el guiÃ³n por la letra acertada en la lista.
            palabra_oculta = " ".join(lista_palabra_usuario)       # Actualizamos la palabra oculta con la nueva letra.
            set_letras_acertadas.add(letra)

        else:
            set_letras_falladas.add(letra)
            vidas -= 1
            actualizar_horca(vidas) # Actualizamos la horca segÃºn las vidas restantes
            print(f"Has fallado...âŒ pierdes una vida. Te quedan {vidas} vidas")
        

        print(f"Letras falladas: {set_letras_falladas}") # Informamos de las vidas restantes y letras falladas 
        seguir_jugando = vidas > 0 and set_letras_acertadas !=set_letras_palabra_secreta 
        print(palabra_oculta)

        # Mostramos la horca actualizada
        for linea in horca:
            print(linea)
        
        # Verificamos si el juego ha terminado
        start = vidas > 0 and set_letras_acertadas != set_letras_palabra_secreta

    # Establecemos las condiciones que determinan la salida del bucle while: que el usuario acierte la palabra o pierda todas sus vidas.
    if vidas == 0:
        print(f"Ohhhhh, has perdido todas tus vidas ğŸ’”. La palabra secreta era: {palabra_secreta}")

    else:
        print (f"ğŸ¥³Â¡ENHORABUENA!ğŸ¥³ Acertaste la palabra secreta: {palabra_secreta}")


    print("\n----------------------SE ACABÃ“ LA PARTIDA----------------------.\n")

    seguir_jugando = input(f"Â¿Otra partida? Teclea si para continuar, y no para abandonar el juego: ").lower()
    if seguir_jugando == 'si':
        continue # Regresa al inicio del bucle while para iniciar una nueva partida, reseteando el contador de vidas a 0
            
    elif seguir_jugando == 'no':
        print("Â¡Nos vemos en la prÃ³xima partida! ğŸ‘‹")
        break # Sale del bucle y termina el juego
    else:
        print("OpciÃ³n no vÃ¡lida, saliendo del juego.ğŸ‘‹")
        break # Anticipamos que el jugador puede introducir una respuesta incorrecta a las dos opciones dadas, y si es asÃ­ manejamos esa entrada como no vÃ¡lida y salimos del juego
```
