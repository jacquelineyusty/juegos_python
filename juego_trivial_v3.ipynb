```python
import random  #Importar libreria random para escoger preguntas aleatoriamente

import unicodedata #Importar esta libreria para eliminar las tildes si el jugador ingresa respuesta con acento

def quitar_tildes(texto):
    # Convertir las letras con tildes a letras sin tildes
    texto_sin_tildes = ''.join(
        c for c in unicodedata.normalize('NFD', texto) #Descomprime los caracteres acentuados, es decir 'a' + acento
        if unicodedata.category(c) != 'Mn'# 'Mn' Es una etiqueta especial que significa "marca no espaciadora", que es una manera técnica de decir "acento" o "diéresis", cosas que se añaden a las letras.
    )
    
    return texto_sin_tildes

#Defiir Banco de preguntas de Geografía
banco_preguntas = { '¿Cuál es el país más pequeño del mundo?' : 'Vaticano', 
                    '¿Qué país tiene más habitantes?' : 'China', 
                    '¿Qué país es el más grande del mundo?' : 'Rusia',
                    '¿Cuál es la montaña más alta del mundo?' : 'Everest', 
                    '¿Cómo se llama el río que atraviesa la selva tropical de Brasil?' : 'Amazonas',
                    '¿Cuál es el río más largo del mundo?' : 'Nilo',
                    "¿Cuál es la capital de Italia?" : "Roma", 
                    "¿En qué ciudad se encuentra la Sagrada Familia?" : "Barcelona", 
                    "¿Qué país tiene la mayor población de África?" : "Nigeria",
                    "¿Cuál es la ciudad más poblada del mundo?" : "Tokio",
                    "¿Cuál es la montaña más alta de España?": "Teide", 
                    "¿Qué río pasa por París?" : "Sena", 
                    "¿Cuál es la capital de Suiza?" : "Berna",
                    "¿Cuál es el continente más poblado?" : "Asia",
                    "¿Cuál es el río que pasa por más paises?" : "Danubio",
                    "¿Cuál es la ciudad con menos habitantes del mundo?" : "Hum",
                    "¿Cuál es el desierto más grande del mundo? " : "Sahara",
                    "¿Cuál es el río que pasa por la ciudad de París?" : "Sena",
                    "¿Qué país se encuentra en dos continentes?" : "Turquia",
                    "¿Cuál es la isla más grande del mundo?" : "Groenlandia",
                    "¿Dónde se encuentra el Machu Pichu?": "Peru",
                    "Cuál es la ciudad más fría del mundo?" : "Oymyakon",
                    "¿Cuál es el río más grande de Colombia?" : "Magdalena",
                    "¿Cuál es el río más largo de la Península Ibérica?" : "Tajo",
                    "¿En qué ciudad de Italia se encuentra la escultura del David de Miguel Angel?" : "Florencia",
                    "¿Cuál es el país más pobre del mundo?": "Burundi",
                    "Los cinco océanos del mundo son Atlántico, Índico, Ártico, Antártico y ____________" : "Pacifico",
                    "¿Dónde se encuentra el volcán más grande la tierra?" : "Hawai",
                    "¿Cuál es el río más largo del Reino Unido?" : "Severn",
                    "¿Cuál es la capital de Canadá?" : "Ottawa",
                    "¿Dónde se encuentra las Grandes Pirámides de Guiza?" : "Egipto",
                    "¿Qué lago tiene una famosa criatura mítica que lleva su nombre?" : "Ness"

                    }


#Dar la bienvenida al jugador
print("**************************************************************************************")
print("                 ¡Bienvenid@ al juego de Trivial de Geografía!")
print("                   Un juego de: TRITURADORAS DE CÓDIGO, S.A.")
print("***************************************************************************************\n")

#Dar la posibilidad al jugador de elegir si quiere ver las instrucciones o empezar a jugar

opcion_usuario = input("Pulsa 'I' si quieres VER LAS INSTRUCCIONES o 'J' para INICIAR EL JUEGO").lower()#Convertimos la respuesta de jugador no tener conflicto

#print("\n ----------------------------------------------------------------------------------\n")
estado = True

while estado:

    if opcion_usuario == 'i':
        print("                     <---INSTRUCCIONES DEL JUEGO--->\n")

        print("- Debes acertar 5 preguntas para ser coronado REY DEL TRIVIAL.✔👍\n" "- Si fallas 3 respuestas recibirás un suspenso en Geografía.❌😧\n""- IMPORTANTE!🚨 Debes responder cada pregunta con una única palabra\n")
        input("Presiona ENTER para comenzar a jugar ")
        opcion_usuario = 'j'
        

    elif opcion_usuario == 'j':
        
        print("-------------------------------------¿PREPARAD@?--------------------------------------\n")
        print("---------------------------------------¿LIST@?----------------------------------------\n")
        print("---------------------------------¡COMIENZA EL JUEGO!----------------------------------\n")
        
        salir_del_programa = False # Este bucle permitirá al jugador seguir encadenando partidas después de ganar o perder. 

        while (not salir_del_programa):
     
            # Iniciar el numero de vidas del jugador. Perderá si llegan a 0.
            vidas = 3

            #Iniciar el numero de aciertos del jugador. Ganará si llegan a 5.
            aciertos = 0

            seguir_jugando = True

            while seguir_jugando :
     
                pregunta, respuesta = random.choice(list(banco_preguntas.items()))         
        
        
                print(f"🚨Atención, pregunta: {pregunta}")
                respuesta_usuario = input(f"Por favor, ingresa tu respuesta").capitalize()
                respuesta_sin_tildes = quitar_tildes(respuesta_usuario)# Instanciamos el método para quitar las tildes
                if respuesta_sin_tildes == respuesta:
                    aciertos += 1              
                    print("¡Respuesta correcta!✔👍") 
                else:
                    vidas -= 1
                    print(f"Respuesta incorrecta ❌😧. La respuesta correcta era: {respuesta}")

                if vidas == 0 or aciertos == 5:# Se continua haciendo preguntas hasta que el jugador gane o pierda
                    seguir_jugando = False
                else: 
                    seguir_jugando = True
                    print(f"📄RECUENTO: Llevas {aciertos} aciertos. Te quedan {vidas} vidas\n")
              
            # El juego sale del bucle por dos motivos: o el jugador ha gastado sus 3 vidas o ha acertado 5 respuestas

            if aciertos == 5:
                print("¡ENHORABUENA! Te coronamos: 🤴REY DEL TRIVIAL DE GEOGRAFÍA.🌏")
            else:
                print("Lo siento!!😢, has perdido todas tus vidas. Te regalamos un suspenso en Geografía🌏: '0' ")
                print("--------------------------------------------------------------------------------------------\n") 
                 

              
            #Dar opcion al jugador de iniciar otra partida o salir del juego
            estado_partida = True
            while estado_partida:
                otra_partida = input("¿Quieres echar otra partida? Pulsa 'J' para JUGAR DE NUEVO o 'S' para SALIR").lower()      
                if otra_partida == 'j':
                    salir_del_programa = False
                    print("¡VAMOS A JUGAR OTRA RONDA!💪")
                    print("-"*90)
                    estado_partida = False
                

                elif otra_partida == 's':
                    salir_del_programa = True
                    print("\n ---------------------------------------------------------------------------------------\n")
                    print("                         ¡HASTA LA PRÓXIMA!🙂")
                    estado_partida = False

                else: 
                    print("Solo puedes elegir 'J' para JUGAR DE NUEVO o 'S' para SALIR")
                    estado_partida = True


        estado = False
    
    else:
        print("Solo puedes elegir 'I' si quieres VER LAS INSTRUCCIONES o 'J' para INICIAR EL JUEGO")
        opcion_usuario = input(" Elige 'I' si quieres VER LAS INSTRUCCIONES o 'J' para INICIAR EL JUEGO").lower()#Convertimos la respuesta de jugador no tener conflicto




    
```
