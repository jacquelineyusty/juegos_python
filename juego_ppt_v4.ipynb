```python
import random

#Saludamos al jugador y le damos la bienvenida.
print("*********************************************************************")
print("            ¡Bienvenid@ al juego piedra 🪨, papel 📄 o tijera ✂️!")
print("               Un juego de: TRITURADORAS DE CÓDIGO, S.A.")
print("*********************************************************************\n")


instrucciones_o_jugar = input("⌨️ Pulsa 'I' para leer las instrucciones 📜o 'J' para empezar a jugar ⏯️⌨️").lower() # Interactuamos con el jugador solicitando respuesta activa para comenzar la partida

if instrucciones_o_jugar == 'i': # Si el jugador escoge 'i' le mostramos las instrucciones.
    print("Recordemos las instrucciones📜:\n"
        
        "El objetivo del juego es vencer al oponente seleccionando el arma que gana según las siguientes reglas: \n"
        
        "- El partida consta de 3 rondas.3️⃣\n"
        "- Puedes elegir un arma distinta por cada ronda.\n"
        "- La piedra aplasta a la tijera.🪨\n"
        "- La tijera corta el papel.✂️\n"
        "- El papel envuelve a la piedra.📄\n"
        "- En caso de que ambos jugadores elijan la misma arma, se produciría un empate y se juega otra vez.🔄\n")
    
    
else: # Si el jugador escoge 'j' (o cualquier otra tecla) arrancamos la partida
    print("----------------------🖱️¿PREPARAD@?🖱️----------------------\n"
          "------------------------🚀¿LIST@?🚀----------------------\n"
          "------------------🎮¡COMIENZA EL JUEGO!🎮----------------------\n\n")

# Establecemos el primer bucle while para permitir que se puedan jugar múltiples partidas
salir_programa = True
while salir_programa:
    contador_rondas = 0

# El juego se desarrolla en un bucle while que se ejecuta mientras que no se hayan jugado más de 3 rondas y que incluye 3 condicionales, donde comparamos ambas tiradas.
    while contador_rondas < 3: 
        arma = input("Elige tu arma para jugar: piedra, papel o tijera.").lower() # Pedimos al usuario que elija su arma, recordándole las opciones, asegurándome que lo que ingrese el usuario sea recibido en minúsculas
        opciones_armas = ["piedra", "papel", "tijera"]
        opcion_aleatoria = random.randint(0,2) 

        if arma not in opciones_armas: # Si el arma dada no es encuentra entre las 3 opciones, mostramos un mensaje de aviso y le damos la oportunidad de escoger de nuevo
            print(f"Has elegido {arma}, esta opción no es válida, elige de nuevo.")
            continue

        if arma in opciones_armas and contador_rondas == 0:
            print("\nRONDA 1")

        if arma in opciones_armas and contador_rondas == 1:
            print("\nRONDA 2")
            
        if arma in opciones_armas and contador_rondas == 2:
            print("\nRONDA 3")

        
        print(f"Has elegido {arma}") # Mostramos al jugador la confirmacion del arma elegida
            
        arma_maquina = opciones_armas[opcion_aleatoria] # La máquina escoge su arma de manera aleatoria
        print(f"El ordenador ha elegido {arma_maquina}") # Mostramos al jugador la opción escogida por la máquina
            
        if arma == "piedra":
            if arma_maquina == "piedra":
                print("¡Has empatado!🥈")
            if arma_maquina == "papel":
                print("¡Has perdido!❌")
            if arma_maquina == "tijera":
                print("¡Has ganado!🥇")
                    
        elif arma == "papel":
            if arma_maquina == "piedra":
                print("¡Has ganado!🥇")
            if arma_maquina == "papel":
                print("¡Has empatado!🥈")
            if arma_maquina == "tijera":
                print("¡Has perdido!❌")

        else:
            if arma_maquina == "piedra":
                print("¡Has perdido!❌")
            if arma_maquina == "papel":
                print("¡Has ganado!🥇")
            if arma_maquina == "tijera":
                print("¡Has empatado!🥈")
            
        contador_rondas +=1

    print("\n----------------------SE ACABÓ LA PARTIDA. ¿TE APETECE OTRA?----------------------.\n")

   #Dar opcion al jugador de iniciar otra partida o salir del juego
    estado_partida = True

    while estado_partida:
            seguir_jugando = input(f"¿Otra partida? Teclea 'J' para seguir jugando o 'S' para salir del juego:").lower()     
            if seguir_jugando == 'j':
                salir_programa = True
                print("¡VAMOS A JUGAR OTRA RONDA!💪")
                print("-"*90)
                estado_partida = False
                

            elif seguir_jugando == 's':
                    salir_programa = False
                    print("\n ---------------------------------------------------------------------------------------\n")
                    print("                     ¡HASTA LA PRÓXIMA!🙂")
                    estado_partida = False
                    

            else: 
                    print("Solo puedes elegir 'J' para JUGAR DE NUEVO o 'S' para SALIR")
                    estado_partida = True
```
